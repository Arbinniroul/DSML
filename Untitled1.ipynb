```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the data
data = {
    "Id": [1, 2, 3, 4, 5],
    "Name": ["Hari", "Curry", "Ram", "Shyam", "Bishal"],
    "Age": [24, 26, 24, 18, None],
    "Location": ["Kathmandu", "Kathmandu", "Butwal", "Pokhara", "Nan"],
    "Salary": [35000, 36000, 53000, np.nan, 50000]
}


df = pd.DataFrame(data)


print("Original DataFrame:")
print(df)


df['Age'].fillna(df['Age'].mean(), inplace=True)


df['Salary'].fillna(df['Salary'].median(), inplace=True)


df['Location'].fillna(df['Location'].mode()[0], inplace=True)

df['Name'].fillna(df['Name'].mode()[0], inplace=True)


print("\nUpdated DataFrame (after replacing NaN values):")
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Salary'], color='skyblue')


plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Bar Graph of Salary by Name')


plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the data
data = {
    "Id": [1, 2, 3, 4, 5],
    "Name": ["Hari", "Curry", "Ram", "Shyam", "Bishal"],
    "Age": [24, 26, 24, 18, None],
    "Location": ["Kathmandu", "Kathmandu", "Butwal", "Pokhara", np.nan],
    "Salary": [35000, 36000, 53000, np.nan, 50000]
}


df = pd.DataFrame(data)


print("Original DataFrame:")
print(df)


df['Location'] = df['Location'].fillna(df['Location'].mode()[0])


df['Salary'] = df['Salary'].fillna(df['Salary'].median())


df['Age'] = df['Age'].fillna(df['Age'].mean())


print("\nUpdated DataFrame (after replacing NaN values):")
print(df)


grouped_df = df.groupby('Location').agg(
    Salary_Mean=('Salary', 'mean'),
    Salary_Median=('Salary', 'median'),
    Salary_Count=('Salary', 'count')
).reset_index()

print("\nGrouped DataFrame based on Location:")
print(grouped_df)

```
